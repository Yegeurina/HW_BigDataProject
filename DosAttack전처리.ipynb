```python
import pandas as pd
import numpy as np
```

```python
label=['TimeStamp','CAN ID','DLC','DATA0','DATA1','DATA2','DATA3','DATA4','DATA5','DATA6','DATA7','Flag']
```

```python
dos = pd.Cov = pd.read_csv('/content/drive/Shareddrives/빅응보 플젝/DoS_dataset.csv',names=label)
```

```python
dos.astype('object')
dos.astype({'TimeStamp' : 'float64'})
dos.astype({'DLC':'int64'})
dos.dtypes
```

```python
# 결측값 있는 행 제거
dos.dropna(axis=0,how='any',inplace=True)
```

```python
dos['CAN ID']=dos['CAN ID'].apply(lambda x : int(str(x),16)) 
for i in range(0,8):
  adr = 'DATA'+str(i)
  #print(adr)
  dos[adr]=dos[adr].apply(lambda x : int(str(x),16))
```

```python
from datetime import datetime
import time
```

```python
import math
```

```python
dos['TimeStamp']=dos['TimeStamp']*math.pow(10,9)
dos['TimeStamp']=pd.to_datetime(dos['TimeStamp'],format="%Y-%m-%d %H:%M:%S")
```

```python
dos['Diff']=dos.TimeStamp.diff()
```

```python
dos = dos.sort_values(by=['CAN ID'],ascending=True)
dos
```

```python
def generate_lagged_features(dos,var,max_lag) : 
  for t in range(1,max_lag+1) :
    dos[var+'_lag'+str(t)]=dos.groupby('CAN ID')[var].shift(t)
```

```python
grouped = dos.groupby('CAN ID')
grouped.size()
Max_num = max(grouped.size())
Min_num = min(grouped.size())
```

```python
generate_lagged_features(dos,'Diff',10) #min max로 설정하니 값이 너무 커서 세션 다운됨... 
dos
```
